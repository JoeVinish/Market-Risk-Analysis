```python
df_returns = np.log(df.drop(['Date','dates'],axis=1)).diff(axis = 0, periods = 1)
```

```python
df_returns.shape
```

```python
df_returns.head(10)
```

```python
df_returns.describe()
```

```python
df_means = df_returns.mean(axis = 0)
df_means
```

```python
df_sd = df_returns.std(axis = 0)
df_sd
```

```python
df_new = pd.DataFrame({'Average':df_means, 'Volatility': df_sd})
df_new
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_new['Volatility'], df_new['Average'], color='blue', marker='o')
plt.title('Scatterplot of Average vs. Volatility')
plt.xlabel('Volatility')
plt.ylabel('Average')
for i, txt in enumerate(df_new.index):
    plt.annotate(txt, (df_new['Volatility'][i], df_new['Average'][i]), textcoords="offset points", xytext=(0,5), ha='center')
plt.show()
```
